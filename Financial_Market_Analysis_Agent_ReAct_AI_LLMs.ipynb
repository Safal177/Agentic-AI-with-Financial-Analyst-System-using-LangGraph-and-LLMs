 # Making Financial Data Tools for ReAct Agent Pipelines


```python
from langchain_core.tools import tool
```

```python
# Need to see stock ticker symbol, company name, Central Index Key (CIK)   # see at dataframe

@tool
def collect_stock_ticker_symbol_company_name_CIK(company_name: str) -> str:
  """
  Collect the stock ticker symbol, company name,and Central Index Key (CIK) for any U.S. company
  with data obtained from SEC filings.  # SEC is for U.S. Securities and Exchange Commission filings.
  """
  company_info = obb.equity.search(company_name, provider = 'sec')
  company_info_table = company_info.to_df().to_markdown()
  return company_info_table
```

```python
from IPython.display import display, Markdown
```

```python
top_companies = ['Nvidia', 'Apple', 'Microsoft', 'Amazon', 'Tesla']

for company_name in top_companies:
    output = collect_stock_ticker_symbol_company_name_CIK.invoke(company_name)
    display(Markdown(f'*** {company_name}'))
    display(Markdown(output))
```

```python
# Company report with fundamental financial analysis

@tool
def fundamental_financial_ratios_growth_metrics(company_ticker_symbol: str) -> str:
    ''' Need to see some financial information about valuation, company growth, profitability, and performance.'''

    # Look at fundamental financial ratios (FMP)
    fundamental_financial_ratios  = obb.equity.fundamental.ratios(symbol = company_ticker_symbol, provider='fmp', period='annual', limit = 7)
    formatted_findamental_financial_ratios_data = fundamental_financial_ratios.to_df().to_markdown()

    # Look at Company growth
    # company_growth = obb.equity.fundamental.growth(symbol = company_ticker_symbol, provider="yfinance", period = "annual", limit = 7)
    # company_growth_metrics = company_growth.to_df().to_markdown()

    # See main financial metrics
    main_financial_metrics = obb.equity.fundamental.metrics(symbol=company_ticker_symbol, provider='yfinance',  period='annual', limit=7)
    main_financial_metrics_data = main_financial_metrics.to_df().to_markdown()

    # Combine fundamental ratios and metrics into a single output.
    fundamental_analysis_report = ('''company ticker symbol:\n''' + company_ticker_symbol + ''':\n\n''' +
              'findamental financial ratios data:\n' + formatted_findamental_financial_ratios_data +
              '\n\nFundamental Metrics:\n' + main_financial_metrics_data)
    return fundamental_analysis_report

```

```python
# Receive stock price information including historical data, the current price quote, and performance metrics for a specified ticker.

from datetime import datetime, timedelta

@tool
def stock_price_information(company_ticker_symbol: str) -> str:
  """
  Receive stock price information including historical data, the current price quote, and performance metrics for a specified ticker.
  """
  # Obtain historical stock price data over the last 12 months with yfinance
  today = datetime.now()
  one_year_ago = (today - timedelta(days=365)).strftime("%Y-%m-%d")
  historical_data = obb.equity.price.historical( provider='yfinance', interval='1d', start_date = one_year_ago, symbol=company_ticker_symbol)
  stock_price_history = historical_data.to_df().to_markdown()

  # Get latest stock price performance statistics using 'fmp' provider
  price_performance_data = obb.equity.price.performance(symbol=company_ticker_symbol, provider='fmp')
  formatted_price_performance_data  = price_performance_data.to_df().to_markdown()

  # Use the 'yfinance' to get the current price quote
  price_quote_data = obb.equity.price.quote(company_ticker_symbol, provider='yfinance')
  formatted_price_quote_data = price_quote_data.to_df().to_markdown()

  # Make a single report from stock price metrics
  stock_summary = (f'company ticker symbol:\n {company_ticker_symbol}'
  f'stock price history:\n{stock_price_history}'
  f'formatted price performance data:\n{ formatted_price_performance_data}'
  f'formatted price quote data:\n{formatted_price_quote_data}')

  return stock_summary

```

```python
result = stock_price_information.invoke('TSLA')
display(Markdown(result))
```

```python
# Collect latest news, articles, and headlines associated with a certain stock

@tool
def current_stock_news(company_ticker_symbol: str) -> str:
  '''Obtain up-to-date news related with a stock ticker'''

  # Get company news
  company_news = obb.news.company(symbol=company_ticker_symbol, provider="yfinance", limit = 12)
  df_company_news = company_news.to_df()

  # Clean news (headline) text
  df_company_news['title'] = (df_company_news['title'].astype(str).str.replace(r'\s+', ' ', regex=True).str.strip())

  return f'latest news: {company_ticker_symbol}\n' + df_company_news[['title']].to_markdown(index=False)

```

```python
result = current_stock_news.invoke('MSFT')
display(Markdown(result))
```

# Overview of the stock market with the most active stocks, top declining stocks, and stocks with steepest gains as well as find latest market trends and stock fluctuations

```python
@tool
def overview_stock_market_data() -> str:
    '''Watching real time market movements data.'''

    # Using yfinance to pull out frequently traded stocks.
    frequently_traded_stocks = obb.equity.discovery.active(limit=12, sort='desc', provider='yfinance')
    frequently_traded_stocks_data = frequently_traded_stocks.to_df().to_markdown()

    # Using yfinance to collect stocks with steepest declines
    stocks_steepest_declines = obb.equity.discovery.losers(limit=12, sort='desc', provider='yfinance')
    stocks_steepest_declines_data = stocks_steepest_declines.to_df().to_markdown()

    # Using yfinance to collect highest gaining stocks.
    highest_gaining_stocks = obb.equity.discovery.gainers(limit=12, sort='desc', provider='yfinance')
    highest_gaining_stocks_data = highest_gaining_stocks.to_df().to_markdown()

    # Combine all market indicators to make a final report.
    response = ('''Compile metrics related with stock market into a display format:\n\n''' +
              'frequently traded stocks data:\n\n' + frequently_traded_stocks_data +
              '\n\n stocks steepest declines data:\n\n' + stocks_steepest_declines_data +
              '\n\n highest gaining stocks data:\n\n' + highest_gaining_stocks_data)
    return response
```

```python
result = overview_stock_market_data.invoke({})
display(Markdown(result))
```

# Making a system prompt for a financial analyst associated with ReAct Agent

```python
financial_market_tools = [collect_stock_ticker_symbol_company_name_CIK, fundamental_financial_ratios_growth_metrics,
         stock_price_information, current_stock_news, overview_stock_market_data]
financial_market_tools
```

```python
from langchain_openai import ChatOpenAI

chatopenai_gpt = ChatOpenAI(model = 'gpt-4o', temperature=0)
```

```python
from langchain_core.messages import SystemMessage
```

```python
agent_system_instructions = '''
You are a financial analyst in artificial intelligence (AI). You are required to provide accurate stocks data and stock market directions.

Find the intent of the user query:
•	As the query is individual stock, you are responsible for finding the ticker symbol and then getting appropriate price and news related to financial information.
•	As the query is macro-market, you should analyze market trends using all possible available market data analysis tools.
•	As the query is ambiguous, you need to communicate in a professional manner with users and request further information.

To get all necessary information, you need to use only the financial tools given by users.
If necessary, you can apply a set of tools. However, please do not perform extraneous tool operations.

You need to show results using markdown formatting and keep a professional style.
You encourage users to perform their own research before making investment decisions.
Before committing to any investment, encourage users to review details personally.
You are responsible for handling special characters such as $ to \$ for accurate layout.
'''
```

```python
system_message_prompt = SystemMessage(content=agent_system_instructions)
```

```python
from langgraph.prebuilt import create_react_agent

stock_analysis_agent = create_react_agent(model=chatopenai_gpt, tools=financial_market_tools, state_modifier=system_message_prompt)
stock_analysis_agent
```

```python
from IPython.display import Image, display

graph = stock_analysis_agent.get_graph()

display(Image(graph.draw_mermaid_png()))
```

# ReAct financial analysis pipeline test


```python
from langchain_core.messages import HumanMessage
```

```python
for update in stock_analysis_agent.stream(
    {'messages': [HumanMessage(content = 'Please provide me with a list of top investment stocks associated with today’s market conditions. Focus on top U.S. companies showing a healthy financial position, solid future growth path and manageable risk. Include detailed reasons for each stock. Before making any investment, I would like to review them. ')]},
    stream_mode='values'):
    update['messages'][-1].pretty_print()

display(Markdown(update['messages'][-1].content))
```

```python
for update in stock_analysis_agent.stream(
    {'messages': [HumanMessage(content = 'Before investing in Amazon, I would like an overview of its recent financial performance and current market trends.')]},
    stream_mode='values'):
    update['messages'][-1].pretty_print()

display(Markdown(update['messages'][-1].content))
```

# ReAct Agent Recommendation

```python
def generate_financial_report(query):
    for output in stock_analysis_agent.stream({'messages': [HumanMessage(content=query)]}, stream_mode = 'values'):
        pass

    display(Markdown(output['messages'][-1].content))
```

```python
generate_financial_report('Evaluate and contrast the profitability trends of Nvidia, Apple, Microsoft, Amazon, and Tesla. Offer investment guidance supported by solid financial logic.')
```
